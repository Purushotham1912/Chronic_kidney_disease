# 3. Finding the missing values

```python
data_set.drop_duplicates(subset=['age','bp','sg','al','su','rbc','pc','pcc','ba','bgr','bu','sc','sod','pot','hemo','pcv','wc','rc','htn','dm','cad','appet','pe','ane','classification'])
```

```python
data_set=data_set.drop_duplicates(subset=['age','bp','sg','al','su','rbc','pc','pcc','ba','bgr','bu','sc','sod','pot','hemo','pcv','wc','rc','htn','dm','cad','appet','pe','ane','classification'])
```

```python
data_set.drop('id',inplace=True,axis=1)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean',missing_values=np.nan)
imputer=imputer.fit(data_set[['age','bp','bgr','bu','sod','sg','al','sc','pot','hemo']])
data_set[['age','bp','bgr','bu','sod','sg','al','sc','pot','hemo']]=imputer.transform(data_set[['age','bp','bgr','bu','sod','sg','al','sc','pot','hemo']])
```

```python
data_set
```

```python
mn.matrix(data_set)
```

```python
mean=data_set['su'].mean()
data_set['su'].fillna(mean,inplace=True)
```

```python
data_set['rbc'].fillna("normal",inplace=True)
data_set['pc'].fillna("normal",inplace=True)
data_set['pcc'].fillna("notpresent",inplace=True)
data_set['ba'].fillna("notpresent",inplace=True)
```
